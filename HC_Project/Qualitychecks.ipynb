```python
%sql
-- Validation: Find departments in Silver that don't exist in Bronze
SELECT 
  s.dept_code,
  s.dept_name,
  s.effective_date,
  s.is_active
FROM edl_hc_datamart.silver.departments s
WHERE NOT EXISTS (
  SELECT 1 
  FROM edl_hc_datamart.bronze.departments b 
  WHERE b.dept_code = s.dept_code
);
```

```python
%sql
-- Validation: Find jobs in Silver that don't exist in Bronze
SELECT 
  s.employee_id,
  s.position_id,
  s.effective_date,
  s.is_active
FROM edl_hc_datamart.silver.jobs s
WHERE NOT EXISTS (
  SELECT 1 
  FROM edl_hc_datamart.bronze.jobs b 
  WHERE b.employee_id = s.employee_id 
    AND b.position_id = s.position_id
);
```

```python
%sql
-- Validation: Find attendance records in Silver that don't exist in Bronze
SELECT 
  s.employee_id,
  s.attendance_date,
  s.effective_date,
  s.is_active
FROM edl_hc_datamart.silver.attendance s
WHERE NOT EXISTS (
  SELECT 1 
  FROM edl_hc_datamart.bronze.attendance b 
  WHERE b.employee_id = s.employee_id 
    AND b.date = s.attendance_date
);
```

```python
%sql
-- Validation: Find employees in Gold that don't exist in Silver (current active records)
SELECT 
  g.employee_key,
  g.first_name,
  g.last_name,
  g.email
FROM edl_hc_datamart.gold.dim_employees g
WHERE NOT EXISTS (
  SELECT 1 
  FROM edl_hc_datamart.silver.employees s 
  WHERE s.employee_id = g.employee_key 
    AND s.is_active = true
);
```

```python
%sql
-- Validation: Find departments in Gold that don't exist in Silver (current active records)
SELECT 
  g.dept_code,
  g.dept_name
FROM edl_hc_datamart.gold.dim_departments g
WHERE NOT EXISTS (
  SELECT 1 
  FROM edl_hc_datamart.silver.departments s 
  WHERE s.dept_code = g.dept_code 
    AND s.is_active = true
);
```

```python
%sql
-- Validation: Find job history records in Gold with orphaned employee or department references
SELECT 
  'Missing Employee' AS validation_type,
  f.employee_key,
  f.dept_key,
  f.job_title
FROM edl_hc_datamart.gold.fact_job_history f
WHERE NOT EXISTS (
  SELECT 1 
  FROM edl_hc_datamart.silver.employees s 
  WHERE s.employee_id = f.employee_key
)

UNION ALL

SELECT 
  'Missing Department' AS validation_type,
  f.employee_key,
  f.dept_key,
  f.job_title
FROM edl_hc_datamart.gold.fact_job_history f
WHERE NOT EXISTS (
  SELECT 1 
  FROM edl_hc_datamart.silver.departments s 
  WHERE s.dept_code = f.dept_key
);
```

```python
%sql
-- Show all sources that had changes in the latest run (records_written > 0)
WITH latest_run AS (
  SELECT MAX(run_start_ts) AS latest_ts
  FROM edl_hc_datamart.audit.audit_ingestion
)
SELECT 
  a.pipeline_name,
  a.layer,
  a.source_name,
  a.target_table,
  a.records_read,
  a.records_written,
  a.last_load_date,
  a.run_start_ts,
  a.run_end_ts,
  a.last_status
FROM edl_hc_datamart.audit.audit_ingestion a
INNER JOIN latest_run lr ON DATE_FORMAT(a.run_start_ts, 'yyyy-MM-dd HH:mm:ss') = DATE_FORMAT(lr.latest_ts, 'yyyy-MM-dd HH:mm:ss')
WHERE a.records_written > 0
ORDER BY a.layer, a.source_name;
```

```python
%sql
-- Show new/changed records in Silver layer from latest run
-- These are records with effective_date matching the latest load_date
WITH latest_load AS (
  SELECT MAX(last_load_date) AS latest_date
  FROM edl_hc_datamart.audit.audit_ingestion
  WHERE layer = 'silver'
)
SELECT 
  'employees' AS table_name,
  employee_id AS record_key,
  first_name,
  last_name,
  is_active,
  effective_date,
  end_date
FROM edl_hc_datamart.silver.employees e
INNER JOIN latest_load ll ON DATE_FORMAT(e.effective_date, 'yyyy-MM-dd') = DATE_FORMAT(ll.latest_date, 'yyyy-MM-dd')

UNION ALL

SELECT 
  'departments' AS table_name,
  dept_code AS record_key,
  dept_name,
  NULL AS last_name,
  is_active,
  effective_date,
  end_date
FROM edl_hc_datamart.silver.departments d
INNER JOIN latest_load ll ON DATE_FORMAT(d.effective_date, 'yyyy-MM-dd') = DATE_FORMAT(ll.latest_date, 'yyyy-MM-dd')

UNION ALL

SELECT 
  'jobs' AS table_name,
  CAST(employee_id AS STRING) AS record_key,
  position_id AS first_name,
  NULL AS last_name,
  is_active,
  effective_date,
  end_date
FROM edl_hc_datamart.silver.jobs j
INNER JOIN latest_load ll ON DATE_FORMAT(j.effective_date, 'yyyy-MM-dd') = DATE_FORMAT(ll.latest_date, 'yyyy-MM-dd')

ORDER BY table_name, record_key;
```

```python
%sql
-- Show records that were deactivated (soft deleted) in the latest run
-- These have end_date set to the latest load_date and is_active = false
WITH latest_load AS (
  SELECT MAX(last_load_date) AS latest_date
  FROM edl_hc_datamart.audit.audit_ingestion
  WHERE layer = 'silver'
)
SELECT 
  'employees' AS table_name,
  employee_id AS record_key,
  first_name,
  last_name,
  is_active,
  effective_date,
  end_date
FROM edl_hc_datamart.silver.employees e
INNER JOIN latest_load ll ON DATE_FORMAT(e.end_date, 'yyyy-MM-dd') = DATE_FORMAT(ll.latest_date, 'yyyy-MM-dd')
WHERE e.is_active = false

UNION ALL

SELECT 
  'departments' AS table_name,
  dept_code AS record_key,
  dept_name,
  NULL AS last_name,
  is_active,
  effective_date,
  end_date
FROM edl_hc_datamart.silver.departments d
INNER JOIN latest_load ll ON DATE_FORMAT(d.end_date, 'yyyy-MM-dd') = DATE_FORMAT(ll.latest_date, 'yyyy-MM-dd')
WHERE d.is_active = false

UNION ALL

SELECT 
  'jobs' AS table_name,
  CAST(employee_id AS STRING) AS record_key,
  position_id AS first_name,
  NULL AS last_name,
  is_active,
  effective_date,
  end_date
FROM edl_hc_datamart.silver.jobs j
INNER JOIN latest_load ll ON DATE_FORMAT(j.end_date, 'yyyy-MM-dd') = DATE_FORMAT(ll.latest_date, 'yyyy-MM-dd')
WHERE j.is_active = false

ORDER BY table_name, record_key;
```
