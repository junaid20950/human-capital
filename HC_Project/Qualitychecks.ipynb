```python
%sql
-- See all Bronze load dates and counts to compare
SELECT 
    'attendance' as table_name,
    business_date,
    _meta_load_date,
    COUNT(*) as record_count
FROM edl_hc_datamart.bronze.attendance
GROUP BY business_date, _meta_load_date

UNION ALL

SELECT 'jobs', business_date, _meta_load_date, COUNT(*)
FROM edl_hc_datamart.bronze.jobs
GROUP BY business_date, _meta_load_date

UNION ALL

SELECT 'employees', business_date, _meta_load_date, COUNT(*)
FROM edl_hc_datamart.bronze.employees
GROUP BY business_date, _meta_load_date

UNION ALL

SELECT 'departments', business_date, _meta_load_date, COUNT(*)
FROM edl_hc_datamart.bronze.departments
GROUP BY business_date, _meta_load_date

ORDER BY table_name, business_date DESC, _meta_load_date DESC;
```

```python
%sql
-- Compare Silver data by effective_date to see what was added when
SELECT 
    'attendance' as table_name,
    DATE(effective_date) as effective_date,
    COUNT(*) as active_count,
    SUM(CASE WHEN is_active = true THEN 1 ELSE 0 END) as currently_active,
    SUM(CASE WHEN is_active = false THEN 1 ELSE 0 END) as inactive_versions
FROM edl_hc_datamart.silver.attendance
GROUP BY DATE(effective_date)

UNION ALL

SELECT 'jobs', DATE(effective_date), COUNT(*),
       SUM(CASE WHEN is_active = true THEN 1 ELSE 0 END),
       SUM(CASE WHEN is_active = false THEN 1 ELSE 0 END)
FROM edl_hc_datamart.silver.jobs
GROUP BY DATE(effective_date)

UNION ALL

SELECT 'employees', DATE(effective_date), COUNT(*),
       SUM(CASE WHEN is_active = true THEN 1 ELSE 0 END),
       SUM(CASE WHEN is_active = false THEN 1 ELSE 0 END)
FROM edl_hc_datamart.silver.employees
GROUP BY DATE(effective_date)

UNION ALL

SELECT 'departments', DATE(effective_date), COUNT(*),
       SUM(CASE WHEN is_active = true THEN 1 ELSE 0 END),
       SUM(CASE WHEN is_active = false THEN 1 ELSE 0 END)
FROM edl_hc_datamart.silver.departments
GROUP BY DATE(effective_date)

ORDER BY table_name, effective_date DESC;
```
